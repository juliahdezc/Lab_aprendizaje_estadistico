**Aguas Frescas**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("Aguas_frescas.xlsx")
df.head()
```

```python
x = df[['KAHLUA', 'COCO', 'HORCHATA', 'MANDARINA', 'TAMAÑO', 'HIELO']]

y = df['CALIF']
```

```python
X_modelo = sm.add_constant(x)
ols = sm.OLS(y, X_modelo)
results = ols.fit()
results.summary()
```

**Carros**

```python
motor = pd.read_excel("Motor Trend Car Road Tests.xlsx")
motor.dropna()
motor.drop_duplicates()
motor.head()
```

```python
x2 = motor.drop(columns=["mpg","model"])
y2 = motor["mpg"]
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x2, y2, train_size = 0.4, random_state = 137)
```

```python
scaler = StandardScaler().fit(Xtrain)
X_scaled = scaler.transform(Xtrain)
```

```python
X_modelo2 = sm.add_constant(X_scaled)
ols2 = sm.OLS(Ytrain, X_modelo2)
results2 = ols2.fit()
results2.summary()
```

**Con test**

```python
X_scaled_test = scaler.transform(Xtest)
```

```python
y_pred = results2.predict(sm.add_constant(X_scaled_test))
r2 = r2_score(Ytest, y_pred)
r2
```
