```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
bh = pd.read_excel("Boston Housing Dataset 1978.xlsx")
bh2 = bh.dropna()
bh2 = bh2.drop_duplicates()
bh2.head()
```

```python
bh2.describe()
```

```python
lr1 = LinearRegression()
```

```python
X = bh2[['CRIM', 'ZN', 'INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
y = bh2[['MEDV']]
```

```python
lr1.fit(X,y)

B0 = lr1.intercept_
B0
```

```python
B1 = lr1.coef_
B1
```

```python
X = sm.add_constant(X)
ols = sm.OLS(y, X)
results = ols.fit()
results.summary()
```

# NOTA

LOS BETAS NO SON COMPARABLES porque los factores no están estandarizados. O sea, los rangos de los valores son muy diferentes 

Por esto, es necesario escalar el dataframe

**Escalamiento con MINMAX**

```python
min_CRIM = np.min(bh2['CRIM'])
bh2["CRIM"] = bh2["CRIM"]- min_CRIM
max_CRIM = np.max(bh2['CRIM'])
bh2["CRIM"] = bh2["CRIM"] / max_CRIM
```

```python
min_ZN = np.min(bh2['ZN'])
bh2["ZN"] = bh2["ZN"]- min_ZN
max_ZN = np.max(bh2['ZN'])
bh2["ZN"] = bh2["ZN"] / max_ZN
```

```python
min_INDUS = np.min(bh2['INDUS'])
bh2["INDUS"] = bh2["INDUS"]- min_INDUS
max_INDUS = np.max(bh2['INDUS'])
bh2["INDUS"] = bh2["INDUS"] / max_INDUS
```

```python
x_escalada = bh2
x_escalada = x_escalada.drop(columns=["MEDV"])

for col in x_escalada.columns:
    x_min = x_escalada[col].min()
    x_escalada[col] = x_escalada[col] - x_min
    x_max = x_escalada[col].max()
    x_escalada[col] = x_escalada[col] / x_max
    
x_escalada.head()
```

```python
X2 = sm.add_constant(x_escalada)
ols2 = sm.OLS(y, X2)
results2 = ols2.fit()
results2.summary()
```
