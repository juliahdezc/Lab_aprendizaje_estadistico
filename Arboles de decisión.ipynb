```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import statsmodels.api as sm
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.stats import t 
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('Advertising.csv')
df
```

## 1

```python
X = df[['TV', 'radio', 'newspaper']]
X = sm.add_constant(X)
y = df['sales']
```

```python
scaler = StandardScaler().fit(X)
x_scaled = scaler.transform(X)
```

```python
modelo=LinearRegression()
modelo.fit(x_scaled,y)
y_pred= modelo.predict(x_scaled)
betas = np.insert(modelo.coef_, 0, modelo.intercept_)
r2=r2_score(y,y_pred)
print("R2",r2)
```

## 2

```python
tree = DecisionTreeRegressor()
tree.fit(x_scaled,y)
y_tree = tree.predict(x_scaled)
r2_tree = r2_score(y, y_tree)
r2_tree
```

## 3

```python
print(f'R^2 de Linear Regression: {r2.round(4)} VS. R^2 de DecisionTreeRegressor: {r2_tree}')
```

**OBSERVACIONES**

A pesar de que ambos modelos tienen un error cuadrado bueno, el del árbol es perfecto

## 4

```python
Xtrain, Xtest, ytrain, ytest=train_test_split(X,y,train_size=0.7, random_state = 42)

Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)
```

```python
modelo4 = LinearRegression()
modelo4.fit(Xtrain_scaled, ytrain)
y_pred_train = modelo4.predict(Xtrain_scaled)
y_pred_test = modelo4.predict(Xtest_scaled)

print("R2 train:", r2_score(ytrain, y_pred_train).round(4))
print("R2 test:",  r2_score(ytest, y_pred_test).round(4))
```

```python
tree4 = DecisionTreeRegressor()
tree4.fit(Xtrain_scaled,ytrain)
y_pred_train4 = tree4.predict(Xtrain_scaled)
y_pred_test4 = tree4.predict(Xtest_scaled)

print("R2 train:", r2_score(ytrain, y_pred_train4).round(4))
print("R2 test:",  r2_score(ytest, y_pred_test4).round(4))
```

**OBSERVACIONES**

Con las variables separadas a X train y test , el R^2 de la regresión lineal da buenos resultados, y con el árbol, pero el R^2 del árbol da perfecto para train, y con un muy buen resultado para test, con 0.94

## 5

```python
PF=PolynomialFeatures(degree = 2)
Xtrain_scaledPF = PF.fit_transform(Xtrain)
Xtest_scaledPF = PF.transform(Xtest)


modeloPF =LinearRegression()
modeloPF.fit(Xtrain_scaledPF,ytrain)
y_predPF_train= modeloPF.predict(Xtrain_scaledPF)
y_predPF_test = modeloPF.predict(Xtest_scaledPF)

print("R2 train:", r2_score(ytrain,y_predPF_train).round(4))
print("R2 test:",  r2_score(ytest, y_predPF_test).round(4))
```

```python
treePF = DecisionTreeRegressor()
treePF.fit(Xtrain_scaledPF,ytrain)
y_predPF_trainTree= treePF.predict(Xtrain_scaledPF)
y_predPF_testTree = treePF.predict(Xtest_scaledPF)

print("R2 train:", r2_score(ytrain, y_predPF_trainTree).round(4))
print("R2 test:",  r2_score(ytest, y_predPF_testTree).round(4))
```

**OBSERVACIONES**

Con las variables de X train y test modificadas con PolynomialFeatures, el R^2 de la regresión lineal da muy parecida, pero con el árbol, pero el R^2 del árbol da perfecto para train, y casi perfecto para test, con 0.97
